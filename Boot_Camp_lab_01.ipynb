# Lab_01 Linear Regression
## Synthetic Dataset


```python
import numpy as np
# Generate 10 random linear data points
x = np.linspace(0, 10, 10)
y = 2 * x + 3 + np.random.randn(10)
# Print the data points
for i in range(len(x)):
    print(f"({x[i]}, {y[i]})")
# Plot the data points
import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.show()
```

```python
import pandas as pd

# Create a Pandas DataFrame from the data
df = pd.DataFrame({
    "x": x,
    "y": y
})

# Print the DataFrame
print(df.to_string())
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 10 random linear data points
x = np.linspace(0, 10, 10)
y = 2 * x + 3 + np.random.randn(10)

# Fit a linear regression model to the data
coefficients = np.polyfit(x, y, 1)

# Calculate the line values
line_x = np.linspace(np.min(x), np.max(x), 100)
line_y = coefficients[0] * line_x + coefficients[1]

# Plot the data points and the line
plt.scatter(x, y)
plt.plot(line_x, line_y, color='red')
plt.show()
```
